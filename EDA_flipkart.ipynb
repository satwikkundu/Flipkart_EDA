```python
# Merge with Customers
df = df_sales.merge(
    df_customers,
    how='left',
    left_on='Customer Name Index',
    right_on='Customer Index'
)
```

```python
# Merge with Products

df = df.merge(
    df_products,
    how='left',
    left_on='Product Description Index',
    right_on='Index'
)
```

```python
# Merge with Regions
df = df.merge(
    df_regions,
    how='left',
    left_on='Delivery Region Index',
    right_on='id'
)
```

## Revenue and profit analysis

### Subtask:
Calculate revenue and profit, and analyze them across different dimensions (products, channels, regions) as per the initial goal.


**Reasoning**:
Calculate the profit column and then group the data by product, channel, and region to calculate total revenue and profit for each dimension. Then, display the top performers for each dimension based on revenue and profit.



```python
df['Profit'] = df['Line Total'] - df['Total Unit Cost']

product_performance = df.groupby('Product Name').agg(
    total_revenue=('Line Total', 'sum'),
    total_profit=('Profit', 'sum')
).sort_values(by='total_revenue', ascending=False)

channel_performance = df.groupby('Channel').agg(
    total_revenue=('Line Total', 'sum'),
    total_profit=('Profit', 'sum')
).sort_values(by='total_revenue', ascending=False)

region_performance = df.groupby('Region').agg(
    total_revenue=('Line Total', 'sum'),
    total_profit=('Profit', 'sum')
).sort_values(by='total_revenue', ascending=False)

print("Top 10 Products by Revenue:")
display(product_performance.head(10))

print("\nTop 10 Products by Profit:")
display(product_performance.sort_values(by='total_profit', ascending=False).head(10))

print("\nChannel Performance by Revenue:")
display(channel_performance)

print("\nChannel Performance by Profit:")
display(channel_performance.sort_values(by='total_profit', ascending=False))

print("\nRegion Performance by Revenue:")
display(region_performance)

print("\nRegion Performance by Profit:")
display(region_performance.sort_values(by='total_profit', ascending=False))
```

## Seasonal trends and outliers

### Subtask:
Analyze the data for seasonal patterns and identify any outliers in sales or profit.


**Reasoning**:
Extract year, month, and day of the week from 'OrderDate', group by year and month to calculate monthly sales and profit, and then plot these to visualize seasonal trends.



```python
df['OrderYear'] = df['OrderDate'].dt.year
df['OrderMonth'] = df['OrderDate'].dt.month
df['OrderDayOfWeek'] = df['OrderDate'].dt.dayofweek

monthly_sales_profit = df.groupby(['OrderYear', 'OrderMonth']).agg(
    total_revenue=('Line Total', 'sum'),
    total_profit=('Profit', 'sum')
).reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales_profit, x=monthly_sales_profit.index, y='total_revenue', label='Total Revenue')
sns.lineplot(data=monthly_sales_profit, x=monthly_sales_profit.index, y='total_profit', label='Total Profit')
plt.title('Monthly Revenue and Profit Over Time')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.xticks(ticks=monthly_sales_profit.index, labels=[f'{y}-{m}' for y, m in zip(monthly_sales_profit['OrderYear'], monthly_sales_profit['OrderMonth'])], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

**Reasoning**:
Group by year and day of the week to calculate weekly sales and profit, and then plot these to visualize daily trends.



**Reasoning**:
The previous plotting code failed because the `label` argument is handled by `hue` in `seaborn.lineplot`. Remove the `label` arguments from the line plots and let seaborn handle the legend using `hue`.



```python
weekly_sales_profit = df.groupby(['OrderYear', 'OrderDayOfWeek']).agg(
    total_revenue=('Line Total', 'sum'),
    total_profit=('Profit', 'sum')
).reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=weekly_sales_profit, x='OrderDayOfWeek', y='total_revenue', hue='OrderYear')
sns.lineplot(data=weekly_sales_profit, x='OrderDayOfWeek', y='total_profit', hue='OrderYear')
plt.title('Daily Revenue and Profit Over Time by Year')
plt.xlabel('Day of Week')
plt.ylabel('Amount')
plt.xticks(ticks=[0, 1, 2, 3, 4, 5, 6], labels=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.legend(title='Metric and Year') # Add a title to the legend
plt.tight_layout()
plt.show()
```

**Reasoning**:
Create box plots for 'Line Total' and 'Profit' to visually identify outliers across the entire dataset.



```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

sns.boxplot(x=df['Line Total'], ax=axes[0])
axes[0].set_title('Box Plot of Line Total (Outliers)')

sns.boxplot(x=df['Profit'], ax=axes[1])
axes[1].set_title('Box Plot of Profit (Outliers)')

plt.tight_layout()
plt.show()

# **Reasoning**:
# Further analyze outliers by creating box plots for 'Line Total' and 'Profit' grouped by 'Product Name', 'Channel', and 'Region'. This allows for identification of dimensions that might be driving the outliers.

plt.figure(figsize=(15, 8))
sns.boxplot(x='Channel', y='Line Total', data=df)
plt.title('Box Plot of Line Total by Channel')
plt.show()

plt.figure(figsize=(15, 8))
sns.boxplot(x='Channel', y='Profit', data=df)
plt.title('Box Plot of Profit by Channel')
plt.show()

plt.figure(figsize=(20, 8))
sns.boxplot(x='Region', y='Line Total', data=df)
plt.title('Box Plot of Line Total by Region')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(20, 8))
sns.boxplot(x='Region', y='Profit', data=df)
plt.title('Box Plot of Profit by Region')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# For products, due to potentially many products, we'll look at the top N products
top_n = 10
top_products_revenue = product_performance.head(top_n).index
df_top_products = df[df['Product Name'].isin(top_products_revenue)]

plt.figure(figsize=(20, 8))
sns.boxplot(x='Product Name', y='Line Total', data=df_top_products)
plt.title(f'Box Plot of Line Total for Top {top_n} Products')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(20, 8))
sns.boxplot(x='Product Name', y='Profit', data=df_top_products)
plt.title(f'Box Plot of Profit for Top {top_n} Products')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# **Insights from Outlier Analysis:**
# - The box plots for 'Line Total' and 'Profit' across the entire dataset show the presence of significant outliers, particularly on the higher end.
# - Analyzing outliers by 'Channel' reveals that some channels might have a wider spread of values and potentially more high-value transactions or costs leading to outliers.
# - Similarly, examining 'Region' indicates that certain regions might have more variability in sales and profit figures, potentially contributing to outliers.
# - Focusing on the top products by revenue also shows varying distributions and potential outliers within specific product categories.
# - Further investigation is needed to understand the nature of these outliers. Are they legitimate large transactions, data entry errors, or something else? Addressing significant outliers might involve data cleaning or a deeper dive into the specific circumstances of those transactions.
```
